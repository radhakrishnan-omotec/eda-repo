# **Project :**  Perceptions and Impacts of the Transition to Electric Vehicles: A Data-Driven Analysis

**ABSTRACT :**
This exploratory data analysis (EDA) delves into the perceptions and expectations surrounding the transition from internal combustion (IC) vehicles to electric vehicles (EVs).

Through a series of structured questions, we investigate demographic characteristics, employment status, commuting behaviors, and attitudes towards EV adoption. Furthermore, we explore concerns about switching to EVs, anticipated impacts on employment levels across various sectors, and willingness to participate in training programs for new skills.

The analysis aims to identify patterns, trends, and potential areas of concern regarding the transition to EVs, offering insights into measures to mitigate negative employment impacts and facilitate a smooth transition towards sustainable transportation.
