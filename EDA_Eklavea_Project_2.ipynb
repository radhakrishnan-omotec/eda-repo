**ABSTRACT :**
This exploratory data analysis (EDA) delves into the perceptions and expectations surrounding the transition from internal combustion (IC) vehicles to electric vehicles (EVs).

Through a series of structured questions, we investigate demographic characteristics, employment status, commuting behaviors, and attitudes towards EV adoption. The analysis aims to identify patterns, trends, and potential areas of concern regarding the transition to EVs, offering insights into measures to mitigate negative employment impacts and facilitate a smooth transition towards sustainable transportation.

**Age groups reveal varying levels of interest in EV adoption, guiding targeted marketing efforts and policy initiatives:**

Age distribution: Analyzing age groups can provide insights into the demographic trends of respondents interested in transitioning to electric vehicles (EVs), highlighting potential target demographics for EV marketing campaigns and policy initiatives.

1) What is your age group?

````python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("/content/eda-repo/Output_Enhanced.csv")

# Check the first few rows of the dataframe
print(df.head())

# Check the columns in the dataframe
print(df.columns)

# Assuming "q1_new" column exists, generate pie chart for its value counts
if 'Q1_new' in df.columns:
    # Count the occurrences of each value in the 'q1_new' column
    q1_counts = df['Q1_new'].value_counts()

    # Plotting the pie chart
    fig, ax = plt.subplots(figsize=(8, 8))
    wedges, labels, autopct = ax.pie(q1_counts, labels=q1_counts.index, autopct='%1.1f%%', startangle=140)

    # Adding legend on top right side
    ax.legend(wedges, q1_counts.index,
              title="Age Groups",
              loc="upper right",
              bbox_to_anchor=(1, 0, 0.5, 1))

    # Adding custom legend labels
    custom_labels = ["0", "1", "2", "3", "4", "5", "6"]
    plt.gca().set_prop_cycle(None)  # Reset color cycle
    for i, custom_label in enumerate(custom_labels):
        plt.plot([], label=custom_label)

    plt.title('Distribution of Age Groups')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
else:
    print("Column 'q1_new' not found in the CSV file.")
````
